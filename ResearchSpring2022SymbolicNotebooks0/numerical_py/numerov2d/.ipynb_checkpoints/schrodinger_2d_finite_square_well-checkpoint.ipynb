```python
from numerov2d import *
```

```python
def finiteSquareWell(
            xPositions : np.ndarray, 
            yPositions : np.ndarray, 
            potentialHeight : float, 
            width : float, 
            height : float
        ) -> np.ndarray: 
    potential : np.ndarray = np.zeros(xPositions.shape)
    potential = np.where((xPositions <= width) | (xPositions >= (1 - width)), potentialHeight, potential)
    potential = np.where((yPositions <= height) | (yPositions >= (1 - height)), potentialHeight, potential)
    return potential
```

```python
pointCount : int = 150
```

```python
xPositions, yPositions = positionGrid(pointCount)
```

```python
potentials : np.ndarray = finiteSquareWell(xPositions, yPositions, 1, .1, .1)
```

```python
energies, waveFunctions = compute2dWaveFunction(pointCount, potentials)
```

```python
probabilityDistributions = waveFunctions ** 2
```
