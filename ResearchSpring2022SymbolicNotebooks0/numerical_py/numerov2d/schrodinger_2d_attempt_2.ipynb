```python
pointCount = 150
```

```python
xPositions, yPositions = np.meshgrid(
        np.linspace(0, 1, pointCount, dtype=float), 
        np.linspace(0, 1, pointCount, dtype=float)
    )
```

```python
def infiniteSquareWell(
            xPositions : np.ndarray, 
            yPositions : np.ndarray
        ) -> np.ndarray: 
    return 0 * xPositions
```

```python
def gaussian(
            positions : np.ndarray, 
            offset : float, 
            variance : float
        ) -> np.ndarray: 
    exponent = (-1.0 / 2.0) * (-(positions - offset) ** 2) / (variance ** 2)
    scalar = 1.0 / (np.sqrt(2.0 * np.pi) * variance)
    return scalar * np.exp(exponent)
    
```

```python
def gaussian2d(
            xPositions : np.ndarray, 
            yPositions : np.ndarray, 
            xOffset : float, 
            yOffset : float, 
            variance : float
        ) -> np.ndarray: 
    return gaussian(xPositions, xOffset, variance) * gaussian(yPositions, yOffset, variance)
```

```python
potentials : np.ndarray = infiniteSquareWell(xPositions, yPositions)
```

```python
energies, waveFunctions = compute2dWaveFunction(pointCount, potentials)
```
