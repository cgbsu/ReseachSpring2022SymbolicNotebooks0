```python
from numerov3d import *
```

```python
from matplotlib import cm
```

```python
pointCount : int = 150
```

```python
grid = makeLinspaceGrid(pointCount, 1, 3)
```

```python
def inifiniteSquareWell(grid : MeshGrid) -> np.ndarray: 
    return 0 * grid.x
```

```python
potential = inifiniteSquareWell(grid)
```

```python
figure = plt.figure(0, figsize=(9, 9))
axis = figure.add_subplot(111, projection = "3d")
axis.scatter3D(grid.x, grid.y, grid.z, c = potential, cmap = cm.seismic, s = 0.001, alpha = .6, antialiased = True)
```

```python
waves = computeWaveFunction(potential)
```

```python
contourCount = 200
```

```python
plt.figure(figsize=(9, 9))
plt.contourf(grid.x, grid.y, waves.waveFunctions[0], contourCount)
```

```python
plt.figure(figsize=(9, 9))
plt.contourf(grid.x, grid.y, waves.probabilities[0], contourCount)
```

```python
plt.figure(figsize=(9, 9))
plt.contourf(grid.x, grid.y, waves.waveFunctions[2], contourCount)
```

```python
plt.figure(figsize=(9, 9))
plt.contourf(grid.x, grid.y, waves.probabilities[2], contourCount)
```
