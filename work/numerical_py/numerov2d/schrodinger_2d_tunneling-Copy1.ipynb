```python
from numerov2d import *
```

```python
def tunnelingCase(
            xPositions : np.ndarray, 
            yPositions : np.ndarray, 
            barrierPosition : float, 
            barrierWidth : float, 
            potentialHeight : float
        ) -> np.ndarray: 
    potentials = np.zeros(xPositions.shape)
    return np.where(
            (xPositions <= (barrierPosition + barrierWidth)) \
                    & (xPositions >= (barrierPosition - barrierWidth)), 
            potentialHeight, 
            potentials
        )
```

```python
pointCount : int = 150
```

```python
xPositions, yPositions = positionGrid(pointCount)
```

```python
potentials : np.ndarray = tunnelingCase(xPositions, yPositions, .6, .01, .001)
```

```python
energies, waveFunctions = compute2dWaveFunction(pointCount, potentials)
```

```python
probabilityDistributions = waveFunctions ** 2
```
