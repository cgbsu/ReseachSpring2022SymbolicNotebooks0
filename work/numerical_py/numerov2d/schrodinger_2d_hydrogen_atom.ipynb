```python
from numerov2d import *
```

```python
def hydrogenAtom(
            xPosition : float, 
            yPosition : float, 
            centerX : float, 
            centerY : float, 
            bottom : float
        ) -> np.ndarray: 
    distance = 1 / (np.sqrt(((xPosition - centerX) ** 2) + ((yPosition - centerY) ** 2)) + bottom)
    return distance
```

```python
pointCount : int = 150
```

```python
xPositions, yPositions = positionGrid(pointCount)
```

```python
potentials : np.ndarray = hydrogenAtom(xPositions, yPositions, .5, .5, 1e-10)
```

```python
plt.imshow(potentials)
```

```python
energies, waveFunctions = compute2dWaveFunction(pointCount, potentials)
```

```python
probabilityDistributions = waveFunctions ** 2
```
