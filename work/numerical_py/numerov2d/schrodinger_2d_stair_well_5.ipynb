```python
from numerov2d import *
```

```python
pointCount : int = 150
```

```python
xPositions, yPositions = positionGrid(pointCount)
```

```python
potentials : np.ndarray = stairwell(xPositions, yPositions, .1, [1.0 / 3.0, 1.0 / 3.0, 1.0 / 3.0], [1.0 / 3.0, 2.0 / 3.0, 1.0])
```

```python
potentials
```

```python
plt.imshow(potentials)
```

```python
energies, waveFunctions = compute2dWaveFunction(pointCount, potentials, EigenValueTypes.HALF_SPECTRUM)
```

```python
probabilityDistributions = waveFunctions ** 2
```
