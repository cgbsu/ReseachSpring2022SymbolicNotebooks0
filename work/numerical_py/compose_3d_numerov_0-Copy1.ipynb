```python
from numerov3d.numerov3d import *
```

```python
import numerov2d.numerov2d as n2d
```

```python
pointCount : int = 150
```

```python
grid = makeLinspaceGrid(pointCount, 1, 2)
```

```python
potential = n2d.gaussian2d(grid.x, grid.y, .5, .6, .1)
```

```python
plt.imshow(potential)
```

```python
waves = computeWaveFunction(potential)
```

```python
contourCount = 200
```

```python
plt.figure(figsize=(9, 9))
plt.contourf(grid.x, grid.y, waves.waveFunctions[0], contourCount)
```

```python
plt.figure(figsize=(9, 9))
plt.contourf(grid.x, grid.y, waves.waveFunctions[2], contourCount)
```
