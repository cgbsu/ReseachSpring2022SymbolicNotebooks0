```python
###################
# Setup ###########
#                 #
#    +-------+    #
#    |       |    #
#    |       |    #
#    |       |    #
#    +_______+    #
#    +  _|_  +    #
#    |       |    #
#    |   O   |    #
#    |   |   |    #
#    +-------+    #
#                 #
###################

###############################################################################
# There are two cylenders of equal mass and radius, they are connected by a ###
# pinion or a pole with a disc inside a disk-shaped cavity inside the bottom ##
# cylender, represented by _|_ in the above block comment picture. There is ###
# also a child cylender, embedded inside the bottom cylender represneted by O #
# in the block comment diagram above. Theory:                               | #
# Spin the two "main" cylenders off of eachother, there can be an armature ####
# in the "top" cylender (perhapse the _|_) and a stator in the "bottom" #######
# cylender. The "top cylender" will hereafter be refered too as the ###########
# "proppeller" and the "bottom cylender" as the "dragger". The "child #########
# cylender" will also be referred too as the "child". After sspinning up the ##
# two main cylenders to an angular velocity omega_M_in, Allow the proppller ###
# to spin freeley, detaching itself momentarily from the dragger. Then spin ###
# the child off of the dragger in the same direction as the dragger to slow ###
# the dragger. Let us say the propellor had a force A + B when it was #########
# detached and now has kinetic energy proportial too that. The dragger had ####
# the same force in the opposite direction. B is the force of gravity on the ##
# system, while A is the offset force or "propellent force." When the two #####
# cylenders make contact inside the radial cavity again and the propeller #####
# stops spinning freely there is a differential in energy between the two #####
# cylenders! Since the dragger has slowed down (and exherted a force in doing #
# so, thus slowing the propeller by this alone "killing" some of its energy ###
# "stored" in linear velocity) the two now slow down to reach equilibrium #####
# but there is a differential in force, a smaller one going down, and a #######
# larger one going up. But they can once again push off one another and #######
# repeat the cycle thus providing propellentless/medium-less thrust!! #########
# This of course provided that A + B - dragger_force > mg so dragger force ####
# must be less than B and must bleed down force below A. Thats got to be one ##
# fast child! #################################################################
###############################################################################
```

```python
torque = lambda force, radial : force.cross( radial )
```

```python
torque_in = torque( main_driver_force * N.k, interface_cavity_radius * N.i )
```

```python
gravitational_force_relation = sp.Eq( gravitational_force, ( ( 2 * main_cylender_mass ) + child_cylender_mass ) * gravitational_acceleration )
```

```python
initial_forces = sp.Eq( sp.Abs( torque_in ), sp.Abs( propel_force + gravitational_force ) )
```

```python
initial_forces
```

```python
child_torque = torque( child_cylender_force * N.k, child_cylender_radius * N.i )
```

```python
child_torque
```

```python
drag_torque = main_cylender_inertia * drag_cylender_angular_velocity 
```

```python
drag_torque
```

```python
drag_cylender_slowdow = sp.Eq( drag_torque, sp.Eq( sp.Abs( torque_in ) - sp.Abs( child_torque ), ( main_cylender_inertia * main_cylender_input_angular_velocity ) - ( child_cylender_input_angular_velocity * child_cylender_inertia ) ), evaluate=False )
```

```python
drag_cylender_slowdow
```

```python
drag_inertial_relation = sp.Eq( drag_torque, ( main_cylender_inertia * main_cylender_input_angular_velocity ) - ( child_cylender_input_angular_velocity * child_cylender_inertia ) )
```

```python
drag_inertial_relation
```

```python
drive_force_transfer_coefficient = sp.Symbol( "\mu_d" )
```

```python
kinetic_friction_coefficient = sp.Symbol( "\mu_k" )
```

```python
u = sp.vector.Vector( N.i / 3, N.j / 3, N.k / 3 )
```

```python
drive_function = sp.Eq( drive( drive_force_transfer_coefficient, kinetic_friction_coefficient ), 
                    sp.Abs( N.k.cross( N.i ) * interface_cavity_radius \
                    * ( ( main_driver_force * drive_force_transfer_coefficient ) \
                    - ( ( differential_torque_relation.rhs ) \
                       / ( interface_cavity_radius - child_cylender_radius ) ) \
                       * kinetic_friction_coefficient )
                ) 
        )
```

```python
drive_function
```

```python
time = sp.Symbol( 't' )
```

```python
phase = sp.pi / 2
```

```python
mu_drive = .9
```

```python
make_mu_kinetic = lambda ratio : mu_drive * ratio
```

```python
mu_kinetic = make_mu_kinetic(  1 / 10 )
```

```python
##########################################################################
# Modeling this as the two drivers, the child and the main driver, being #
# out of phase, when one stops, the otehr kicks in, modeling it sort of ##
# as a friction ##########################################################
##########################################################################

make_test_function = lambda mu_drive, mu_kinetic, phase : ( drive_function.subs( 
        drive_force_transfer_coefficient, mu_drive * sp.sin( time ) ).subs( 
                kinetic_friction_coefficient, 
                mu_kinetic * sp.cos( time + phase ) 
            ).rhs.subs( N.j, 1 ) ).subs( 
        sp.Abs( main_driver_force * interface_cavity_radius ), 
        main_driver_force * interface_cavity_radius ).subs( 
                child_cylender_force * child_cylender_radius, 
                child_cylender_force * child_cylender_radius
            ).refine()
test_function = make_test_function( mu_drive, mu_kinetic, phase )
#help( test_function )
test_function = test_function.args[ 0 ]
```
